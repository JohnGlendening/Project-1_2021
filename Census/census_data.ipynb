```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# File to Load (Remember to Change These)
file1="census_data.csv"

# Read Purchasing File and store into Pandas data frame

df= pd.read_csv(file1)
```

```python
df.head()
```

```python
Texas=df[["Label","Texas!!Estimate"]]
Texas
```

```python
#Places enthinicities and estimates in new data frame.
#grouped all ethnicities row 7 and down in multicultural 

race_ethnicity=["White","Black or African American","American Indian and Alaska Native","Asian","Native Hawaiian or Pacific Islander","Multicultural"]
Estimate=[21273326,3553922,144993,1446073,27026,4998109]

Texas_df=pd.DataFrame({"Ethnicity":race_ethnicity,
                       "Population Estimate":Estimate
                       })

Texas_df
```

```python
#Pie chart of the percentage(%) of population in Texas
sizes=Texas_df["Population Estimate"]

labels=Texas_df["Ethnicity"].unique()

explode= (0,0,0.2,0,0.5,0.05)

plt.pie(sizes, labels=labels,explode=explode,
        autopct="%1.1f%%", shadow=True, startangle=90)

plt.title("Census Percentage of Population for Ethnicities in texas")
```

```python
Ethincities = Texas_df["Ethnicity"]
Cases_load = Texas_df["Population Estimate"]
plt.title("Census Population per Ethnicity in Texas")
plt.ylabel("Number of People (per million)")
plt.xlabel("Ethnicity")
width = 0.5 
plt.xticks(rotation=90)
plt.bar(Ethincities, Cases_load, width)
plt.tick_params(axis="x", which="major", labelsize=12)
plt.show()
```
