```python
Texas=df[["Label","Texas!!Estimate"]]

race_ethnicity=["White","Black or African American","American Indian and Alaska Native","Asian","Native Hawaiian or Pacific Islander","Multicultural"]
Estimate=[21273326,3553922,144993,1446073,27026,4998109]

Texas_df=pd.DataFrame({"Ethnicities":race_ethnicity,
                       "Population Estimates":Estimate
                       })

Texas_df
```

```python
Texas_Spread =Texas_df["Ethnicities"]
Texas_estimate = Texas_df["Population Estimates"]
plt.title("Number of cases per Ethnicity")
plt.ylabel("Number of cases (per million)")
plt.xlabel("Ethnicity")
width = 0.5 
plt.xticks(rotation=90)
plt.bar(Texas_Spread, Texas_estimate, width)
plt.tick_params(axis="x", which="major", labelsize=12)
plt.show()
```
