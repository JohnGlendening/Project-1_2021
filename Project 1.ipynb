```python
import matplotlib.pyplot as plt
import pandas as pd

#data files
moderna_vac_file = "Project 1 data/COVID-19_Vaccine_Distribution_Allocations_by_Jurisdiction_-_Moderna (1).csv"
Pfizer_vac_file = "Project 1 data/COVID-19_Vaccine_Distribution_Allocations_by_Jurisdiction_-_Pfizer (1).csv"

#read data 
moderna_vac = pd.read_csv(moderna_vac_file)
pfizer_vac = pd.read_csv(Pfizer_vac_file)
```

```python
moderna_vac.head()
```

```python
#shows there are 13 non-states in df
moderna_vac.nunique()
```

```python
#Create clean data frame for moderna
clean_moderna_df_1 = moderna_vac[moderna_vac["Jurisdiction"] != "American Samoa"]
clean_moderna_df_2 = clean_moderna_df_1[clean_moderna_df_1["Jurisdiction"] != "District of Columbia"]
clean_moderna_df_3 = clean_moderna_df_2[clean_moderna_df_2["Jurisdiction"] != "Federal Entities"]
clean_moderna_df_4 = clean_moderna_df_3[clean_moderna_df_3["Jurisdiction"] != "Guam"]
clean_moderna_df_5 = clean_moderna_df_4[clean_moderna_df_4["Jurisdiction"] != "Mariana Islands"]
clean_moderna_df_6 = clean_moderna_df_5[clean_moderna_df_5["Jurisdiction"] != "Marshall Islands"]
clean_moderna_df_7 = clean_moderna_df_6[clean_moderna_df_6["Jurisdiction"] != "Micronesia"]
clean_moderna_df_8 = clean_moderna_df_7[clean_moderna_df_7["Jurisdiction"] != "New York City"]
clean_moderna_df_9 = clean_moderna_df_8[clean_moderna_df_8["Jurisdiction"] != "Palau"]
clean_moderna_df_10 = clean_moderna_df_9[clean_moderna_df_9["Jurisdiction"] != "Puerto Rico"]
clean_moderna_df_11 = clean_moderna_df_10[clean_moderna_df_10["Jurisdiction"] != "U.S. Virgin Islands"]
clean_moderna_df_12 = clean_moderna_df_11[clean_moderna_df_11["Jurisdiction"] != "Philadelphia"]
clean_moderna_df = clean_moderna_df_12[clean_moderna_df_12["Jurisdiction"] != "Chicago"]

#shows only states are in df
clean_moderna_df.nunique()
```

```python
#show first and second dose allocations are exactly the same
clean_moderna_df.describe()
```

```python
jurisdiction = clean_moderna_df["Jurisdiction"]
Dose_Allocations = clean_moderna_df["1st Dose Allocations"]
plt.title("Vaccine Dose Allocation Per State (Moderna)")
plt.ylabel("Number of Doses Allocated")
plt.xlabel("State")
width = 0.5 
plt.xticks(rotation=90)
plt.bar(jurisdiction, Dose_Allocations, width)
plt.tick_params(axis="x", which="major", labelsize=7)
plt.show()
```

```python
plt.title("Vaccine Dose Allocation Per State (Moderna)")
plt.ylabel("Number of Doses Allocated")
plt.xlabel("State")
plt.xticks(rotation=90)
plt.tick_params(axis="x", which="major", labelsize=7)
plt.grid(True)
plt.scatter(jurisdiction, Dose_Allocations)
```
